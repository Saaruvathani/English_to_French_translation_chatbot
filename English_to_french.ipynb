```python
from transformers import pipeline
```

Importing Helsinki-NLP/opus-mt-en-fr from hugging face

```python
translator = pipeline("translation_en_to_fr" , model="Helsinki-NLP/opus-mt-en-fr")
```

Importing modules to add a streamline function


```python
import time
```

Adding streamline function


```python
def stream(text):
  print('Bot :',end='')
  words=text.split()
  for word in words :
    print(word,end=' ',flush=True)
    time.sleep(0.1)
  print()
```

Building the chatbot with dynamic typing effect

```python
print("Welcome to your Translator Chatbot! Type 'quit' to exit.")
while True:
  user_input = input("You: ")
  if user_input.lower()=='quit':
    print("Bot : Bye for now !")
    break
  response = translator(user_input)
  stream(response[0]['translation_text'])
```
